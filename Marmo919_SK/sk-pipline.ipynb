```python
from os import listdir
from os.path import isfile, join
import h5py
import cv2
import numpy as np
from skimage.util import view_as_blocks
from time import gmtime, strftime
from multiprocessing import Pool
from os import mkdir
```

```python
def GetBlockList(overlap=10):
    x1=0
    y1=512
    blocklist=[]
    count = 0
    while (x1+512<36000):
        x2=0
        y2=512
        while (x2+512<48000):
            blocklist.append(np.asarray([x1,y1,x2,y2,count]))
            count = count + 1
            x2=y2-overlap
            y2=x2+512
        x1=y1-overlap
        y1=x1+512
    return blocklist
```

```python
def GetCoord(data,overlap=10):
    x=(data[3]+2*overlap)/512-1
    y=(data[1]+2*overlap)/512-1
    return [x,y]
```

```python
def Order2Cood(a):
    #assuming x is 0 to 94, y is 0 to 70
    x=a%95
    y=(a-(a%95))/95
    return [x,y]
```

```python
def Cood2Order(data):
    #assuming x is 0 to 92, y is 0 to 68
    return (data[1]) * 93 + data[0]
    #
```

```python
def blockprocess(img,metadata,filecount):
    #print type(metadata)
    sub=img[metadata[0]:metadata[1],metadata[2]:metadata[3],:]
    Gmask= np.max(sub,axis=2) != sub[:,:,1]
    sub[Gmask]=0
    sub[sub[:,:,1]<50] =0
    sub=cv2.cvtColor(sub,cv2.COLOR_BGR2GRAY)
    cv2.imwrite(str(metadata[4])+'/'+str(filecount)+'.jp2',sub)
```

```python
def FileProcess(filemeta):
    #read image
    testfile=filemeta[0]
    print 'processing '+testfile
    count=filemeta[1]
    data=h5py.File(testfile)
    img=np.asarray(data['img'],'uint8')
    img=img.T
    #get blocklist
    blocks=GetBlockList()
    for k in blocks:
        blockprocess(img,k,count)
    print 'finishing '+testfile
```

```python
#here start the main script:
count=0
with open('MATdata/orderfile.txt') as f:
    content = f.readlines()
#remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content]     
para=[]
for i in content:
    para.append(('MATdata/'+i,count))
    count = count + 1
```

```python
p=Pool(10)
p.map(FileProcess,para)
```
