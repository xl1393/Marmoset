```python
def GetBlockList(overlap=10):
    x1=0
    y1=512
    blocklist=[]
    count = 0
    while (x1+512<36000):
        x2=0
        y2=512
        while (x2+512<48000):
            blocklist.append(np.asarray([x1,y1,x2,y2,count]))
            count = count + 1
            x2=y2-overlap
            y2=x2+512
        x1=y1-overlap
        y1=x1+512
    return blocklist
```

```python
def GetCoord(data,overlap=10):
    x=(data[3]+2*overlap)/512-1
    y=(data[1]+2*overlap)/512-1
    return [x,y]
```
